# Lion Nose Color

Can you estimate a lion's age from a photograph of its nose?

Lions are born with pink noses that gradually darken with age, developing black freckles until they eventually turn completely black.

The dataset `lion-nose-color.csv` contains data on 91 lions from the Serengeti, each monitored by researchers since birth. Because their exact ages are known, researchers can analyze photographs to determine the percentage of black on their noses and assess how well nose coloration predicts age.

```python
import pandas as pd
import matplotlib.pyplot as plt
from linear_model import LinearModel

df = pd.read_csv('lion-nose-color.csv')
df.head()
```

# Project Ideas

- Create a linear regression model to predict `age` from `proportion_black`

- Use the model to estimate a lion's age when its nose is 50% black. 

#### Considerations

```python
plt.scatter(df['age'],df['proportion_black'])
plt.xlabel('Lion age')
plt.ylabel('Black proportion on nose')
```

```python
# Linear model API
#Create a linear regression model to predict `age` from `proportion_black`

plt.scatter(df['age'],df['proportion_black'])
plt.xlabel('Lion age')
plt.ylabel('Black proportion on nose')
linear = LinearModel()
linear.fit(df['age'], df['proportion_black'])
linear.plot_model(df['age'].min(), df['age'].max(), color="black")
linear.print_model_info()
```

```python
age_at_50 = linear.predict(0.5)
print(f"Geschätztes Alter bei 50% schwarz: {age_at_50:.1f} Jahre")
```

If there are few or no values at 1.0 in the data, a prediction would be based on 100% extrapolation and therefore unreliable.
Biologically, the relationship can saturate non-linearly (the nose can turn completely black, with no further change as it ages), so caution is advised when interpreting the results.
