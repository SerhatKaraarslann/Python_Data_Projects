### Making sense of the data

As you are cleaning a dataframe, it's important to understand what the values in each row and column mean.

Here each row represents one coffee enthusiast and their yes/no answers to each survey question. If the enthusiast adds a particular type of dairy to their coffee, that column will show a value of 1, while a value of 0 indicates that they do not use that type of dairy.

Notice that there are also several `NaN` values (Not a Number), which represents missing values. In this case, NaNs indicate that the survey question was not answered.

Let's use the `isna()` method to take a closer look at how many `NaN` values each column has.

Activity Goals:
- Create a boolean dataframe with True for NaNs and False otherwise.
- Add up the True values per column to get the number of NaNs.


```python
dairy.isna().sum()
```

Notice that each dairy option has the same number of NaN values. Why might this be the case?

This particular survey had a question that asks if you add any dairy to your coffee. If the respondent answered "No", the dairy part of the survey was skipped entirely.

Since these rows have no data, we can drop them using `dropna()`.

Activity Goals:
- Remove rows with NaN values.
- Display the remaining rows.


```python
dairy = dairy.dropna()
dairy #display dataframe
```

After dropping all the rows with NaNs, there are now 523 rows left in the dataframe. Many rows were removed, such as the row with index 1.

Now that we have cleaned the data, we can investigate dairy preferences.

### What dairy options are most preferred?

What percentage of people consume each type of dairy? We can use the `mean()` method on the dataframe to calculate the ratio of "yes" answers for each column. Multiplying by 100 will give us these ratios as percentages.

Activity Goals:
- Calculate the percentage for each dairy.
- Display the result.

```python
dairy_preferences = dairy.mean() * 100
dairy_preferences #display results
```

It looks like whole milk is the most popular. A quick scan of these numbers tells us that the percentages add up to more than 100%. This happens because people often like more than one type of dairy in their coffee.

### Visualizing dairy preferences
Let's visualize the dairy preferences as a horizontal bar chart. We need to first sort `dairy_preferences` and then will use the `barh()` method from matplotlib.

Activity Goals:
- Sort diary_preferences.
- Display the series.

```python
dairy_preferences = dairy_preferences.sort_values()
dairy_preferences #display series
```

And now we can create our horizontal bar plot.

Activity Goals:
- Import matplotlib.
- Make a horizontal bar chart of dairy preferences.
- Label the x-axis.

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6, 3))
plt.barh(dairy_preferences.index, dairy_preferences)
plt.xlabel("Percent")
```

The most popular plant based dairy choice is oat milk. This might be a great choice for the default dairy option in a specialty coffee shop.

## Research Questions for Complete Dataset Analysis

Based on the comprehensive coffee survey data, we'll investigate these key business questions:

1. **Customer Segmentation**: What are the distinct coffee consumer profiles based on age, spending habits, and preferences?

2. **Location Strategy**: Where do different demographic groups prefer to drink coffee, and what does this mean for coffee shop placement?

3. **Premium Product Opportunities**: What demographic factors correlate with higher spending and willingness to pay premium prices?

4. **Equipment vs. Cafe Spending**: How does home coffee equipment investment relate to cafe spending patterns?

5. **Expertise & Preferences**: How does self-reported coffee expertise correlate with specific preferences and spending behavior?

Let's start by exploring the dataset structure and key variables.

```python
# Load the complete dataset first
full_survey = pd.read_csv('coffee-survey-full-dataset.csv')
print(f"Full dataset shape: {full_survey.shape}")
print(f"Number of columns: {full_survey.shape[1]}")
print(f"Number of respondents: {full_survey.shape[0]}")

# Explore key demographic and spending columns
key_columns = [
    'What is your age?',
    'How many cups of coffee do you typically drink per day?',
    'In total, much money do you typically spend on coffee in a month?',
    'What is the most you\'ve ever paid for a cup of coffee?',
    'What is the most you\'d ever be willing to pay for a cup of coffee?',
    'Approximately how much have you spent on coffee equipment in the past 5 years?',
    'Lastly, how would you rate your own coffee expertise?',
    'Gender',
    'Education Level',
    'Employment Status'
]

# Check which columns exist in our dataset
available_columns = [col for col in key_columns if col in full_survey.columns]
print("Available key columns:")
for col in available_columns:
    print(f"- {col}")

print(f"\nDataset columns (first 20):")
print(list(full_survey.columns[:20]))
```

```python
# Research Question 1: Customer Segmentation by Age and Consumption
print("=== AGE DEMOGRAPHICS AND COFFEE CONSUMPTION ===")

# Clean and analyze age data
age_col = 'What is your age?'
cups_col = 'How many cups of coffee do you typically drink per day?'

if age_col in full_survey.columns:
    age_data = full_survey[age_col].dropna()
    print(f"\nAge distribution:")
    age_counts = age_data.value_counts().sort_index()
    print(age_counts)
    
    # Visualize age distribution
    plt.figure(figsize=(10, 6))
    age_counts.plot(kind='bar', color='brown', alpha=0.7)
    plt.title('Coffee Survey Participants by Age Group')
    plt.xlabel('Age Group')
    plt.ylabel('Number of Participants')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
else:
    print("Age column not found in expected format")
```

```python
# Research Question 2: Location Preferences Analysis
print("=== WHERE PEOPLE DRINK COFFEE ===")

# Find location preference columns
location_cols = [col for col in full_survey.columns if 'Where do you typically drink coffee?' in col and '(' in col]
print(f"Location preference columns found: {len(location_cols)}")

if location_cols:
    # Create a clean location preference dataset
    location_data = full_survey[location_cols].copy()
    
    # Rename columns for clarity
    location_mapping = {}
    for col in location_cols:
        if '(At home)' in col:
            location_mapping[col] = 'At Home'
        elif '(At the office)' in col:
            location_mapping[col] = 'At Office'
        elif '(On the go)' in col:
            location_mapping[col] = 'On the Go'
        elif '(At a cafe)' in col:
            location_mapping[col] = 'At Cafe'
        elif '(None of these)' in col:
            location_mapping[col] = 'Other'
    
    location_data = location_data.rename(columns=location_mapping)
    
    # Remove rows with all NaN values
    location_clean = location_data.dropna(how='all')
    
    # Calculate percentages for each location
    location_preferences = location_clean.mean() * 100
    location_preferences = location_preferences.sort_values(ascending=False)
    
    print("Coffee consumption location preferences:")
    print(location_preferences.round(1))
    
    # Visualize location preferences
    plt.figure(figsize=(10, 6))
    location_preferences.plot(kind='barh', color='darkgreen', alpha=0.7)
    plt.title('Where Do Coffee Enthusiasts Typically Drink Coffee?')
    plt.xlabel('Percentage of Respondents (%)')
    plt.tight_layout()
    plt.show()
else:
    print("Location preference columns not found")
```

```python
# Research Question 3: Premium Product Opportunities - Spending Analysis
print("=== COFFEE SPENDING PATTERNS ===")

# Find spending-related columns
spending_cols = [col for col in full_survey.columns if 'money' in col.lower() or 'paid' in col.lower() or 'willing to pay' in col.lower() or 'equipment' in col.lower()]
print("Spending-related columns:")
for col in spending_cols:
    print(f"- {col}")

# Analyze monthly spending if available
monthly_spending_col = 'In total, much money do you typically spend on coffee in a month?'
max_paid_col = 'What is the most you\'ve ever paid for a cup of coffee?'
willing_pay_col = 'What is the most you\'d ever be willing to pay for a cup of coffee?'
equipment_spending_col = 'Approximately how much have you spent on coffee equipment in the past 5 years?'

for col in [monthly_spending_col, max_paid_col, willing_pay_col, equipment_spending_col]:
    if col in full_survey.columns:
        data = full_survey[col].dropna()
        print(f"\n{col}:")
        print(f"Sample values: {data.head().tolist()}")
        if len(data) > 0:
            value_counts = data.value_counts().head(10)
            print(f"Top responses: {value_counts.to_dict()}")
    else:
        print(f"\n{col}: Column not found")
```

```python
# Research Question 4: Coffee Expertise and Preferences
print("=== COFFEE EXPERTISE ANALYSIS ===")

expertise_col = 'Lastly, how would you rate your own coffee expertise?'
strength_col = 'How strong do you like your coffee?'
roast_col = 'What roast level of coffee do you prefer?'
caffeine_col = 'How much caffeine do you like in your coffee?'

# Analyze coffee expertise levels
if expertise_col in full_survey.columns:
    expertise_data = full_survey[expertise_col].dropna()
    print(f"\nCoffee expertise distribution:")
    expertise_counts = expertise_data.value_counts()
    print(expertise_counts)
    
    # Visualize expertise levels
    plt.figure(figsize=(10, 6))
    expertise_counts.plot(kind='pie', autopct='%1.1f%%', colors=['lightcoral', 'gold', 'lightblue', 'lightgreen'])
    plt.title('Self-Reported Coffee Expertise Levels')
    plt.ylabel('')
    plt.show()

# Analyze taste preferences
preference_cols = [strength_col, roast_col, caffeine_col]
for col in preference_cols:
    if col in full_survey.columns:
        data = full_survey[col].dropna()
        if len(data) > 0:
            print(f"\n{col}:")
            value_counts = data.value_counts().head(5)
            print(value_counts)
```

```python
# Research Question 5: Purchase Behavior Analysis
print("=== COFFEE PURCHASE BEHAVIOR ===")

# Find purchase location columns
purchase_cols = [col for col in full_survey.columns if 'purchase coffee' in col and '(' in col]
print(f"Purchase location columns found: {len(purchase_cols)}")

if purchase_cols:
    # Create clean purchase preference dataset
    purchase_data = full_survey[purchase_cols].copy()
    
    # Rename columns for clarity
    purchase_mapping = {}
    for col in purchase_cols:
        if 'National chain' in col:
            purchase_mapping[col] = 'National Chain (Starbucks, Dunkin)'
        elif 'Local cafe' in col:
            purchase_mapping[col] = 'Local Cafe'
        elif 'Drive-thru' in col:
            purchase_mapping[col] = 'Drive-thru'
        elif 'Specialty coffee shop' in col:
            purchase_mapping[col] = 'Specialty Coffee Shop'
        elif 'Deli or supermarket' in col:
            purchase_mapping[col] = 'Deli/Supermarket'
        elif 'Other' in col:
            purchase_mapping[col] = 'Other'
    
    purchase_data = purchase_data.rename(columns=purchase_mapping)
    purchase_clean = purchase_data.dropna(how='all')
    
    # Calculate purchase preferences
    purchase_preferences = purchase_clean.mean() * 100
    purchase_preferences = purchase_preferences.sort_values(ascending=False)
    
    print("Where people purchase coffee on-the-go:")
    print(purchase_preferences.round(1))
    
    # Visualize purchase preferences
    plt.figure(figsize=(12, 6))
    purchase_preferences.plot(kind='bar', color='orange', alpha=0.7)
    plt.title('Coffee Purchase Locations When On-the-Go')
    plt.xlabel('Purchase Location')
    plt.ylabel('Percentage of Respondents (%)')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
else:
    print("Purchase preference columns not found")
```

```python
# Advanced Analysis: Cross-tabulation of Expertise vs Spending Behavior
print("=== EXPERTISE vs SPENDING CORRELATION ===")

expertise_col = 'Lastly, how would you rate your own coffee expertise?'
monthly_spending_col = 'In total, much money do you typically spend on coffee in a month?'
equipment_col = 'Approximately how much have you spent on coffee equipment in the past 5 years?'

# Create a subset with relevant columns for correlation analysis
analysis_cols = [col for col in [expertise_col, monthly_spending_col, equipment_col] if col in full_survey.columns]

if len(analysis_cols) >= 2:
    correlation_data = full_survey[analysis_cols].dropna()
    
    if len(correlation_data) > 10:  # Ensure we have enough data
        print(f"Analysis based on {len(correlation_data)} complete responses")
        
        # Display sample data structure
        print("\nSample data:")
        print(correlation_data.head())
        
        # If we have expertise data, create cross-tabulation
        if expertise_col in correlation_data.columns:
            for spending_col in [monthly_spending_col, equipment_col]:
                if spending_col in correlation_data.columns:
                    print(f"\n=== {expertise_col} vs {spending_col} ===")
                    crosstab = pd.crosstab(correlation_data[expertise_col], 
                                         correlation_data[spending_col])
                    print(crosstab)
    else:
        print("Insufficient data for correlation analysis")
else:
    print("Key columns not available for correlation analysis")

# Summary statistics for key metrics
print("\n=== SUMMARY STATISTICS ===")
for col in analysis_cols:
    if col in full_survey.columns:
        data = full_survey[col].dropna()
        print(f"\n{col}:")
        print(f"Total responses: {len(data)}")
        if len(data) > 0:
            print(f"Unique values: {data.nunique()}")
            print(f"Most common: {data.mode().iloc[0] if len(data.mode()) > 0 else 'N/A'}")
```
